```python
#check for negative valus on GHI, DNI, and DHI and drop these values
colomns=['GHI', 'DNI', 'DHI']
data1 = data[(data[colomns] >=0).all(axis=1)]
data1.shape
```

```python
#checking outliers on colomns ModA and ModB
# Calculate quantiles

Q1 = data1[['ModA', 'ModB']].quantile(0.25)
Q3 = data1[['ModA', 'ModB']].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers from specific columns (ModA and ModB)
data2 = data1[~((data1[['ModA', 'ModB']] < (Q1 - 1.5 * IQR)) | (data1[['ModA', 'ModB']] > (Q3 + 1.5 * IQR))).any(axis=1)]
data2.head(5)
```

**Z-Score Analysis to Flag Outliers**
The Z-score measures how many standard deviations a data point is from the mean. It is particularly useful for detecting outliers.

```python
# Function to calculate Z-scores
def calculate_z_scores(data2, cols):
    z_scores = {}
    for col in cols:
        mean = data2[col].mean()
        std = data2[col].std()
        z_scores[col] = (data2[col] - mean) / std
    return pd.DataFrame(z_scores)

# Calculate Z-scores
columns_to_analyze = ['GHI', 'DNI', 'DHI']
z_scores_data2 = calculate_z_scores(data2, columns_to_analyze)

# To add Z-scores to the original DataFrame
data3 = pd.concat([data2, z_scores_data2.rename(columns=lambda x: f'{x}_zscore')], axis=1)

# Flag outliers (|Z| > 3)
outliers = (z_scores_data2.abs() > 3).any(axis=1)

print("Data with Z-scores:")
print(data3.head(5))

print("\nFlagged Outliers:")
print(data3[outliers].head(5))
```
