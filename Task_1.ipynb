```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#check for negative valus on GHI, DNI, and DHI and drop these values
colomns=['GHI', 'DNI', 'DHI']
data1 = data[(data[colomns] >=0).all(axis=1)]
data1.shape
```

```python
#checking outliers on colomns ModA and ModB
# Calculate quantiles

Q1 = data1[['ModA', 'ModB']].quantile(0.25)
Q3 = data1[['ModA', 'ModB']].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers from specific columns (ModA and ModB)
data2 = data1[~((data1[['ModA', 'ModB']] < (Q1 - 1.5 * IQR)) | (data1[['ModA', 'ModB']] > (Q3 + 1.5 * IQR))).any(axis=1)]
data2.head(5)
```

```python
#to ensure the Timestamp colomn is in datetime format
data2['Timestamp'] = pd.to_datetime(data2['Timestamp'])
```

```python
# #grouping Data by month
data2['month'] = data2['Timestamp'].dt.month

monthly_data = data2.groupby('month')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()
```

```python
#grouping Data by Hour
data2['hour'] = data2['Timestamp'].dt.hour
hourly_data = data2.groupby('hour')[['GHI', 'DNI', 'DHI', 'Tamb']].mean()

```

```python
# 1. Plot Monthly Trends (Bar Chart)
monthly_data.plot(kind='bar', figsize=(12, 6), alpha=0.8)
plt.title('Monthly Averages of GHI, DNI, DHI, and Tamb')
plt.ylabel('Average Values')
plt.xlabel('Month')
plt.legend(title='legend')
plt.show()
```
